```python
def find_outliers_IQR(data):
    q1 = data.quantile(0.25)# First quartile (25th percentile)
    q3 = data.quantile(0.75)# Third quartile (75th percentile)
    IQR = q3 - q1# Interquartile range

    # Find outliers (values outside the range [Q1 - 1.5*IQR, Q3 + 1.5*IQR])
    outliers = data[((data < (q1 - 1.5 * IQR)) | (data > (q3 + 1.5 * IQR)))]
    return outliers
```

```python
# Remove age outliers from the dataset

# Remove outliers from the 'Age' column using IQR method and drop those rows from the dataset
age = find_outliers_IQR(data['Age']).index
data.drop(index=age, inplace=True)

# Remove outliers from the 'Survival_Months' column using IQR method and drop those rows from the dataset
survival_months = find_outliers_IQR(data['Survival_Months']).index
data.drop(index=survival_months, inplace=True)

# Remove outliers from the 'Tumor_Size' column using IQR method and drop those rows from the dataset
tumar_size = find_outliers_IQR(data['Tumor_Size']).index
data.drop(index=tumar_size, inplace=True)

# Remove outliers from the 'Regional_Node_Examined' column using IQR method and drop those rows from the datase
node_examined = find_outliers_IQR(data['Regional_Node_Examined']).index
data.drop(index=node_examined, inplace=True)

# Remove outliers from the 'Reginol_Node_Positive' column using IQR method and drop those rows from the dataset
node_postive = find_outliers_IQR(data['Reginol_Node_Positive']).index
data.drop(index=node_postive, inplace=True)
```
